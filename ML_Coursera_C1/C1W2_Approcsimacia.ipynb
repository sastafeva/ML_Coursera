```python
import numpy as np
%matplotlib inline
import scipy as sp
from matplotlib import pylab as plt
```

```python
# f(x) = sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)
```

```python
# Построение графика заданной функции
x = np.arange(1, 16, 0.1) # задаем массив значений, чтобы график построить
y = np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
plt.plot(x, y)
plt.show()
```

```python
# Вычисление значений заданной функции при подстановке x
def my_func(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
# Проверка работы функции my_func
y1 = np.sin(1 / 5) * np.exp(1 / 10) + 5 * np.exp(-1 / 2)
y2 = np.sin(15 / 5) * np.exp(15 / 10) + 5 * np.exp(-15 / 2)
print(y1, y2)
y1 = my_func(1)
y2 = my_func(15)
print(y1, y2)
```

```python
# Создание матрицы многочлена первой степени для вычисления коэффициентов. y = ax+b. X = 1 и X = 15
m1 = np.array([[1., 1.], [15., 1.]])
v1 = np.array([my_func(1), my_func(15)])
a1, b1 = np.linalg.solve(m1, v1)
print (a1, b1)
```

```python
# Построение графика функции многочлена первой степени
x = np.arange(1, 15, 0.1) # задаем массив значений, чтобы график построить
y = a1*x + b1
plt.plot(x, y)
plt.show()
```

```python
# Создание матрицы многочлена второй степени для вычисления коэффициентов. y = ax2+bx+c. X = [1,8,15]
x = [1,8,15]
m1 = np.array([[x[0]**2, x[0], 1.], [x[1]**2, x[1], 1.], [x[2]**2, x[2], 1.]])
v1 = np.array([my_func(x[0]), my_func(x[1]), my_func(x[2])])
a2, b2, c2 = np.linalg.solve(m1, v1)
print (a2, b2, c2)
print(m1[1])
```

```python
# Построение графика функции многочлена второй степени
x = np.arange(1, 15, 0.1) # задаем массив значений, чтобы график построить
y = a2*(x**2) + b2*x + c2
plt.plot(x, y)
plt.show()
```

```python
# Создание матрицы многочлена третьей степени для вычисления коэффициентов. y = ax3+bx2+cx+d. X = [1,4,10,15]
x = [1, 4, 10, 15]
m3 = np.zeros((4, 4))
for i in range(len(x)):
    m3[i] = [x[i]**3, x[i]**2, x[i], 1.]
v3 = np.array([my_func(i) for i in x])

list_const = np.linalg.solve(m3, v3)
print (list_const)
```

```python
# Построение графика функции многочлена третьей степени
x = np.arange(1, 15, 0.1) # задаем массив значений, чтобы график построить
y = list_const[0]*(x**3) + list_const[1]*x**2 + list_const[2]*x + list_const[3]
plt.plot(x, y)
plt.show()
```

## Построение совместных графиков разных многочленов

```python
# Многочлен первой степени
x = np.arange(1, 15, 0.1) # задаем массив значений, чтобы график построить
y = a1*x + b1

# Многочлен второй степени
h = np.arange(1, 15, 0.1) # задаем массив значений, чтобы график построить
q = a2*(h**2) + b2*h + c2
# Многочлен третьей степени
n = list_const[0]*(x**3) + list_const[1]*x**2 + list_const[2]*x + list_const[3]

# Исходная функция
z = np.arange(1, 16, 0.1) # задаем массив значений, чтобы график построить
f = np.sin(z / 5) * np.exp(z / 10) + 5 * np.exp(-z / 2)
plt.plot(x, y)
plt.plot(z, f)
plt.plot(h, q)
plt.plot(x, n)
plt.show()
```
