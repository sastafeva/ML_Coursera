#### **Решение**

$p-value$ - это достигаемый уровень значимости или вероятность получить такие или еще более экстремальные значения статистики при условии, что верна нулевая гипотеза $H_0$

$\alpha$ - это уровень значимости или порог. Если $p-value$ не достигает $\alpha$, т.е. $p-value \leq \alpha$, нулевая гипотеза отвергается в пользу альтернативы $H_1$. 

$p-value \leq \alpha$ не означает, что $H_1$ верна. Всегда остается вероятность совершить ошибку I-рода, когда отвергается верная нулевая гипотеза.

### Задача 3

Чтобы сравнить распределение сосен с равномерным, посчитайте значение статистики хи-квадрат для полученных 5х5 квадратов. Округлите ответ до двух знаков после десятичной точки.

#### **Решение**

```python
# Построим гистограмму распределения сосен по секторам с наложением равномерного распределения
pylab.figure(figsize=(8,8))
pylab.bar(range(25), pylab.ravel(trees_count.statistic[0]), color = 'g', label = 'Observed frequences')
pylab.bar(range(25), mean_num, color = 'b', alpha = 0.3, label = 'Expected frequences')
pylab.legend()
;
```

$H_0$: сосны растут равномерно, выборка из равномерного распределения.

$H_1$: сосны растут неравномерно, выборка из какого-то другого распределения.

Статистика критерия хи-квадрат: $$\chi^2=\sum_{i=1}^K \frac{\left(n_i- np_i\right)^2}{np_i}$$
При справедливости нулевой гипотезы имеет распределение хи-квадрат с числом степеней свободы $K-1-m$, где $m$ - число параметров распределения, оцененных по выборке.

```python
chisquare_stat = stats.chisquare(pylab.ravel(trees_count.statistic[0]), mean_num, ddof = 0)
chisquare_stat
```

$P-value$ существенно меньше ноля, значит можно смело отклонить нулевую гипотезу о равномерном распределении сосен

```python
print('Значение статистики критерия Хи-квадрат = ', round(chisquare_stat[0], 2))
```

Посчитайте достигаемый уровень значимости.  

Если вы используете функцию $scipy.stats.chi2.cdf$, в качестве значения параметра df нужно взять 24 (это число степеней свободы); если функцию $scipy.stats.chisquare$  — параметр $ddof$ нужно брать равным 0 (это как раз количество параметров теоретического распределения, оцениваемых по выборке).

Отвергается ли гипотеза равномерности на уровне значимости 0.05?

#### **Решение**

```python
# Возьмем из прошлой задачи
print('p_value = ', chisquare_stat[1])
print('Гипотезу о равномерности на уровне значимости 0.05 можно уверенно отвергнуть')
```

Другой вариант

```python
# Построим распределение Хи-квадрат с 24 степенями свободы
x = stats.chi2.rvs(24, size=1000)

pylab.figure(figsize=(8,8))
pylab.hist(x, align = 'mid', density = True, histtype='stepfilled')
;
```

```python
# Посчитаем значение статистики chi2 по приведенной выше формуле
real_trees = pylab.ravel(trees_count.statistic[0])
chi = pylab.sum((real_trees - mean_num)**2 / mean_num)
chi
```

```python
# Посчитаем вероятность получения значений статистики chi или более экстремальных
p_val = stats.chi2.sf(chi, 24)
print('p_value = ', p_val)
print('Гипотезу о равномерности на уровне значимости 0.05 можно уверенно отвергнуть')
```

В обоих случаях ожидаемо получено одинаковое значение p_value, которое практически равно нулю. На основании этого можно с уверенностью утверждать, что сосны растут неравномерно.
